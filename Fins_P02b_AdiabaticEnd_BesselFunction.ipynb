### Step 0: Input data
$\text{Data: } \alpha_A,\alpha_B,R_1,R_2,R_3,H,e_a,N_a,T_A,T_B,\lambda_t,\lambda_a,q_v.$ \\
$\text{Geometric and physical parameters: }$  
\begin{align*}
S_1 &\! = 2 \pi R_1 H \\
S_2 &\! = 2 \pi R_2 H \\
V_t &\! = \pi (R_2^2 - R_1^2) H \\
A_a &\! = 2 \pi (R_3^2 - R_2^2) \\
S_{a2} &\! = 2 \pi R_2 e_a \\
S_{a3} &\! = 2 \pi R_3 e_a \\
m_a &\! = \sqrt{\frac{2 \alpha_B}{\lambda_a e_a}} \\
\phi &\! = \left(\frac{R_3}{R_2} - 1\right) \left(1 + 0.35 \log\left(\frac{R_3}{R_2}\right)\right) \\
\eta_a &\! = \frac{\tanh(m_a R_2 \phi)}{m_a R_2 \phi}
\end{align*}


### Step 1: Temperature and Heat Flux

**Tube:**

\begin{align*}
T_t(r) & = -\frac{\dot{q}_v}{4 \lambda_t} \left( r^2 - R_1^2 \right) + \left( \frac{(T_2 - T_1) - \left( -\frac{\dot{q}_v (R_2^2 - R_1^2)}{4 \lambda_t} \right)}{\ln \left( \frac{R_2}{R_1} \right)} \right) \ln \left( \frac{r}{R_1} \right) + T_1 \\
\dot{q}_t(r) & = -\lambda_t \left( -\frac{\dot{q}_v r}{2 \lambda_t} + \frac{1}{r} \left( \frac{(T_2 - T_1) - \left( -\frac{\dot{q}_v (R_2^2 - R_1^2)}{4 \lambda_t} \right)}{\ln \left( \frac{R_2}{R_1} \right)} \right) \right)
\end{align*}

**Circular Fin:**

\begin{align*}
T_a(r) & = T_B + C_3 I_0(m_a r) + C_4 K_0(m_a r) \\
\dot{q}_a(r) & = -\lambda_a m_a \left( C_3 I_1(m_a r) - C_4 K_1(m_a r) \right)
\end{align*}

### Step 2: Energy Balances.

\begin{align*}
0 & = \{ \dot{q}_t(R_1) S_1 \} - \{ \alpha_A (T_A - T_1) S_1 \} + \{ 0 \}, \quad \text{Eq.(1), 2 unknows ($T_1$,$T_2$).} \\
\end{align*}
\begin{align*}
0 & = \{ \alpha_A (T_A - T_1) S_1 \}  - \left\{ \alpha_B (T_2 - T_B)(S_2 - N_a S_{a2}) + N_a \dot{Q}_{a2} \right\} + \{ \dot{q}_v V_t \}, \quad \text{Eq.(2), 3 unknows ($\dot{Q}_{a2}$).} \\
\end{align*}
\begin{align*}
\dot{Q}_{a2} & = \dot{q}_a(R_2) S_{a2}, \quad \text{Eq.(3), 5 unknows ($C_3$,$C_4$).}  \\
\end{align*}
\begin{align*}
T_2 & =  T_a(R_2), \quad \text{Eq.(4), 5 unknows.}  \\
\end{align*}
\begin{align*}
0 & = \{ \dot{Q}_{a3} \} - \{ 0 \} + \{0\} , \quad \text{Eq.(5), 6 unknows ($\dot{Q}_{a3}$).}  \\
\end{align*}
\begin{align*}
\dot{Q}_{a3} & =  \dot{q}_a(R_3) S_{a3}, \quad \text{Eq.(6), 6 unknows.}  \\
\end{align*}
\begin{align*}
T_3 & =  T_a(R_3) , \quad \text{Eq.(7), 7 unknows ($T_3$)} .
\end{align*}
System of algebraic equations:
\begin{align*}
F_1(T_1,T_2)=0, \\
F_2(T_1,T_2,\dot{Q}_{a2})=0, \\
F_3(\dot{Q}_{a2},C_3,C_4)=0,\\
F_4(T_2,C_3,C_4)=0,\\
F_5(\dot{Q}_{a3})=0,\\
F_6(\dot{Q}_{a3},C_3,C_4)=0,\\
F_7(T_3,C_3,C_4)=0,\\
\end{align*}

### Step 3: Verification.  Global Energy Balance.
\begin{align*}
\{\text{Accumulation rate}\} = \{\text{Input velocity}\}-\{\text{Output velocity}\} +\{\text{Volumetric generation}\}
\end{align*}
\begin{align*}
\{0\} & = \{ \alpha_A (T_A-T_1) S_1 \} - \left\{ \alpha_B (T_2-T_B) ( S_2-N_a S_{a2}) + N_a \left( \int_{A_a} \alpha_B \, (T_a(r)-T_B) \, dA_a \, + \dot{Q}_{a3} \right)  \right\} + \{ \dot{q}_v V_t \},
\end{align*}
\begin{align*}
\mathrm{where } \, \, \dot{Q}_{a3}=0 \, \, \text{ for a fin with adiabatic end},
\, \, \, dA_a=4 \pi r dr.
\end{align*}




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from scipy.special import iv, kv
from scipy.integrate import quad

# STEP 0: Input parameters
#===
alphaA  = 3000  # W/m2 K
alphaB  = 3000  # W/m2 K
R1      = 0.013  # m
R2      = 0.015  # m
R3      = 0.075  # m
H       = 2.3  # m
ea      = 0.004  # m
Na      = 163
TA      = 90  # C
TB      = 25  # C
lambdat = 58.0  # W/(m C)
lambdaa = 58.0  # W/(m C)
qv      = 1e9  # W/m3
#===

#===
S1   = 2 * np.pi * R1 * H
S2   = 2 * np.pi * R2 * H
Vt   = np.pi * (R2**2 - R1**2) * H  # m3
Aa   = 2 * np.pi * (R3**2 - R2**2)
Sa2  = 2 * np.pi * R2 * ea
Sa3  = 2 * np.pi * R3 * ea
ma   = np.sqrt(2 * alphaB / (lambdaa * ea))
phi  = ((R3 / R2) - 1) * (1 + 0.35 * np.log(R3 / R2))
etaa = np.tanh(ma * R2 * phi) / (ma * R2 * phi)
#===

#STEP 1:  T(r) and q(r) for the tube and circular fin
# Tube
def fTt(r,T1,T2):
  Tt = (-qv / (4 * lambdat)) * ( r**2-R1**2 ) + (((T2 - T1) - (- qv * (R2**2 - R1**2) / (4 * lambdat))) / (np.log(R2 / R1))) * np.log(r / R1) + T1
  return Tt

def fqcond_t(r,T1,T2):
  qt = -lambdat * ( (-qv * r / (2 * lambdat))  + (1/r)*(((T2 - T1) - (-qv * (R2**2 - R1**2) / (4 * lambdat))) / (np.log(R2 / R1))) )
  return qt

# Circular fin
def fTa(r,C3,C4):
  Ta = TB + C3 * iv(0, ma * r) + C4 * kv(0, ma * r)
  return Ta

def fqcond_a(r,C3,C4):
  qa = -lambdaa * ma * (C3 * iv(1, ma * r) - C4 * kv(1, ma * r))
  return qa

#STEP 2:  Energy balances
# Numerical resolution
def FunTermotFin(x):
    T1  = x[0]
    T2  = x[1]
    Qa2 = x[2]
    C3  = x[3]
    C4  = x[4]
    Qa3 = x[5]
    T3  = x[6]

    F = np.zeros(7)
    F[0] = fqcond_t(R1,T1,T2) - alphaA * (TA - T1) #Energy balance at the surface S1
    F[1] = alphaA * (TA - T1) * S1 + qv * Vt - alphaB * (T2 - TB) * (S2 - Na * Sa2) - Na * Qa2 #Energy balance in the volume enclosed by surfaces S1-S2
    F[2] = Qa2 - fqcond_a(R2,C3,C4) * Sa2 #Definition
    F[3] = T2 - fTa(R2,C3,C4) #It is assumed Ta(R2)=T2
    F[4] = Qa3 - 0 #Energy balance at Sa3.  Adiabatic end
    F[5] = Qa3 - fqcond_a(R3,C3,C4) * Sa3
    F[6] = T3 - fTa(R3,C3,C4) #Definition
    return F

# Initial guess
x0 = [TA, TB, 0, 1, 1, 0, TB]  # Estimated values [T1,T2,Qa2,C3,C4,Qa3,T3]
x = fsolve(FunTermotFin, x0)

#STEP 3:  Verification
# Verification
T1, T2, Qa2, C3, C4, Qa3, T3 = x
Qv     = qv * Vt
Qconv1 = alphaA * (TA - T1) * (2 * np.pi * R1 * H)
Qconv2 = alphaB * (T2 - TB) * (2 * np.pi * R2 * H - Na * Sa2)
Q2     = Qconv2 + Na * Qa2

# Call fQconv_a after defining it
def Qconvar_dAa(r):
    return alphaB * (fTa(r,C3,C4) - TB) * 2 * 2 * np.pi * r

Qconva, _ = quad(Qconvar_dAa, R2, R3)
Qconva   += Qa3

Res_GlobEnergyBal = Qconv1 - (Qconv2 + Na*Qconva) + Qv  # Should be equal to zero

#Fin efficiency:
FinEff    = Qconva/(alphaB*(T2-TB)*Aa)

# Displaying results
def printReport():
    print("==============================================================================")
    print("REPORT:")
    print(f"T1      = {T1:.6f} [C]")
    print(f"T2      = {T2:.6f} [C]")
    print(f"T3      = {T3:.6f} [C]")
    print(f"Qv      = {Qv:.6f} [W]")
    print(f"Qconv1  = {Qconv1:.6f} [W]")
    print(f"Q2      = {Q2:.6f} [W]")
    print(f"Qconv2  = {Qconv2:.6f} [W]")
    print(f"Qa2     = {Qa2:.6f} [W]")
    print(f"Qa3     = {Qa3:.6f} [W]")
    print(f"Qconva  = {Qconva:.6f} [W].  Qa2 should be equal to Qconva")
    print(f"Na      = {Na:.6f}")
    print(f"Na·Qa2  = {Na * Qa2:.6f} [W]")
    print(f"Global energy balance:")
    print(f"Res_GlobEnergyBal = Qconv1 - (Qconv2 + Na*Qconva) + Qv = {Res_GlobEnergyBal:.6f} [W]")
    print(f"Fin efficiency:")
    print(f"FinEff  = {FinEff:.6f}")
    print(f"FinEff  = {etaa:.6f} (Approximated formula, adiabatic end)")
    print("==============================================================================")

# Print the report
printReport()

# Postprocessing - Temperature profiles
drt = (R2 - R1) / 1000
rt  = np.arange(R1, R2, drt)
Tt  = fTt(rt,T1,T2)

# Circular fin
dra = (R3 - R2) / 1000
ra  = np.arange(R2, R3, dra)
Ta  = fTa(ra,C3,C4)

# TUBE AND CIRCULAR FIN
plt.figure(figsize=(6, 3))
plt.plot(rt, Tt, '-', linewidth=2, color='k', label='Tube Temperature')
plt.plot(ra, Ta, '-', linewidth=2, color='r', label='Fin Temperature')
plt.xlabel('r [m]')
plt.ylabel('T [C]')
plt.title('TUBE AND CIRCULAR FIN')
plt.legend()
plt.grid()
plt.axis('tight')
plt.show()

# Tube only
plt.figure(figsize=(6, 3))
plt.plot(rt, Tt, '-', linewidth=2, color='k')
plt.xlabel('r [m]')
plt.ylabel('T [C]')
plt.title('TUBE')
plt.grid()
plt.axis('tight')
plt.show()
```
